#### Make a calculator by using GUI.

```python
from tkinter import *
import math

root = Tk()

root.config(bg='white')

exp = ''

def press(num):
    global exp
    exp = exp + num
    inp.set(exp)
    
def result():
    global exp
    try:
        s = eval(exp)
    except:
        inp.set('Error')
    else:
        inp.set(s)
    
    finally:
        exp=''
        
def clear_():
    global exp
    exp = ''
    inp.set(exp)
    
def cross_():
    global exp
    exp = exp[:-1]
    inp.set(exp)
        
inp = StringVar()
inp.set('')

ent = Entry(root,textvariable=inp,width=30,font=('Times',15,'bold'))
ent.grid(row=0,columnspan=4)

clear = Button(root,height=2,width=10,text='Clear',bg='black',fg='white',command=clear_)
clear.grid(row=1,column=0)

cross = Button(root,height=2,width=10,text='X',bg='black',fg='white',command= cross_)
cross.grid(row=1,column=1)

remainder = Button(root,height=2,width=10,text='%',bg='black',fg='white',command = lambda : press('%'))
remainder.grid(row=1,column=2)

div = Button(root,height=2,width=10,text='/',bg='black',fg='white',command= lambda : press('/'))
div.grid(row=1,column=3)

bt7 = Button(root,height=2,width=10,text='7',bg='black',fg='white',command= lambda : press('7'))
bt7.grid(row=2,column=0)

bt8 = Button(root,height=2,width=10,text='8',bg='black',fg='white',command= lambda : press('8'))
bt8.grid(row=2,column=1)

bt9 = Button(root,height=2,width=10,text='9',bg='black',fg='white',command= lambda : press('9'))
bt9.grid(row=2,column=2)

mul = Button(root,height=2,width=10,text='*',bg='black',fg='white',command= lambda : press('*'))
mul.grid(row=2,column=3)
                                                                                           
bt4 = Button(root,height=2,width=10,text='4',bg='black',fg='white',command= lambda : press('4'))
bt4.grid(row=3,column=0)

bt5 = Button(root,height=2,width=10,text='5',bg='black',fg='white',command= lambda : press('5'))
bt5.grid(row=3,column=1)

bt6 = Button(root,height=2,width=10,text='6',bg='black',fg='white',command= lambda : press('6'))
bt6.grid(row=3,column=2)
         
sub = Button(root,height=2,width=10,text='-',bg='black',fg='white',command= lambda : press('-'))
sub.grid(row=3,column=3)                                                                                        
                                                                                           
bt1 = Button(root,height=2,width=10,text='1',bg='black',fg='white',command= lambda : press('1'))
bt1.grid(row=4,column=0)

bt2 = Button(root,height=2,width=10,text='2',bg='black',fg='white',command= lambda : press('2'))
bt2.grid(row=4,column=1)

bt3 = Button(root,height=2,width=10,text='3',bg='black',fg='white',command= lambda : press('3'))
bt3.grid(row=4,column=2)

add = Button(root,height=2,width=10,text='+',bg='black',fg='white',command= lambda : press('+'))
add.grid(row=4,column=3)

dot = Button(root,height=2,width=10,text='.',bg='black',fg='white',command= lambda : press('.'))
dot.grid(row=5,column=0)
         
bt0 = Button(root,height=2,width=10,text='0',bg='black',fg='white',command= lambda : press('0'))
bt0.grid(row=5,column=1)

ddiv = Button(root,height=2,width=10,text='//',bg='black',fg='white',command= lambda : press('//'))
ddiv.grid(row=5,column=2)
         
equal = Button(root,height=2,width=10,text='=',bg='black',fg='white',command=result)
equal.grid(row=5,column=3)

root.mainloop()
```
